```python
import openpyxl
from datetime import datetime

wb = openpyxl.Workbook()
ws = wb.active
ws.title = "請求書"

today = datetime.now()
today_str = today.strftime("%Y/%m/%d")
today_filename = today.strftime("%Y%m%d")

ws["B2"] = "請求書"

ws["B4"] = "㈱ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today_str

ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

ws["B11"] = "商品A"
ws["C11"] = 2
ws["D11"] = 10000
ws["E11"] = "=C11*D11"

ws["B12"] = "商品B"
ws["C12"] = 1
ws["D12"] = 15000
ws["E12"] = "=C12*D12"

ws["E13"] = "=SUM(E11:E12)"

ws["B15"] = "小計"
ws["E15"] = "=E13"

ws["B16"] = "消費税"
ws["E16"] = "=E15*0.1"

ws["B17"] = "合計"
ws["E17"] = "=E15+E16"

filename = f"請求書_{today_filename}.xlsx"
wb.save(filename)

print("作成しました:",filename)
```
