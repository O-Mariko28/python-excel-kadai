```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Alignment

# ---------- Step3: 読み込み → 連結 → 集計（pandas） ----------

files = [
    ("2022_年間売上表.xlsx", 2022),
    ("2023_年間売上表.xlsx", 2023),
]

dfs = []
for fname, year in files:
    df = pd.read_excel(fname, sheet_name="Sheet1")

    # もし「売上年」列が無ければ、ファイル年を入れる
    if "売上年" not in df.columns:
        df["売上年"] = year

    dfs.append(df)

all_df = pd.concat(dfs, ignore_index=True)
print(all_df.columns)


# 集計：商品×売上年ごとの合計金額
# ※「金額」列名が違う場合はここを合わせてね（例：売上 など）
summary = (
    all_df
    .groupby(["商品", "売上年"], as_index=False)["金額（千円）"]
    .sum()
    .rename(columns={"金額（千円）": "合計金額（千円）"})
)

# ---------- Step4: 新規Excelに書き込み（pandas） ----------
out_file = "売上集計表.xlsx"
summary.to_excel(out_file, index=False)

# ---------- Step4: ヘッダー書式変更（openpyxl） ----------
wb = load_workbook(out_file)
ws = wb.active

header_fill = PatternFill("solid", fgColor="F2F2F2")
header_font = Font(bold=True)
header_align = Alignment(horizontal="center", vertical="center")

# 1行目（ヘッダー）を装飾
for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = header_align

wb.save(out_file)
print("作成しました：", out_file)
print(summary)

output_file = "売上集計表.xlsx"
summary.to_excel(output_file, index=False)

from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Alignment

# ファイルを開く
wb = load_workbook(output_file)
ws = wb.active

# 薄グレー設定（#F2F2F2）
header_fill = PatternFill("solid", fgColor="F2F2F2")
header_font = Font(bold=True)
header_align = Alignment(horizontal="center")

# 1行目（ヘッダー）を装飾
for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = header_align

# 保存
wb.save(output_file)

print("完成しました：", output_file)

```
